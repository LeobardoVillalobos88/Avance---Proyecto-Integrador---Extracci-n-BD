```python
# Importar librerias
from sklearn.neighbors import KNeighborsClassifier
```

```python
x = [10,4,6,7,7,6,8,9,2,5,6,5,3,2,2,1,8,9,2,7]
y = [9,5,6,7,8,7,6,9,1,5,7,6,2,1,5,5,9,10,4,6]
clase = [1,2,1,1,1,2,2,1,3,3,3,2,3,3,2,2,1,1,3,3]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, c = clase)
plt.show()
```

```python
data = list(zip(x,y))
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(data, clase)
```

```python
new_x = 7
new_y = 5
new_point=[(new_x,new_y)]
predecir = knn.predict(new_point)
```

```python
plt.scatter(x + [new_x] ,y + [new_y], c=clase + [predecir[0]])
plt.text(x=new_x - 0.7, y = new_y-0.7, s=f"Clase: {predecir[0]}")
plt.show()
```

## Support Vector Machine

```python
import pandas as pd
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
modelo = SVC(kernel='linear')
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
print(y_pred)
```
